```python
import requests
import json
import pandas as pd

url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"


querystring = {"ingr": "champagne"}

headers = {
    'X-RapidAPI-Key': "46911d94abmshd89182fc4bc397dp1bde84jsn4da6e2a7559b",
    "X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"
}

dico = requests.get(url, headers=headers, params=querystring).json()
next_page = dico['_links']['next']['href']
values = dico['hints']
while True:
    try:
        dico = requests.get(next_page).json()
        values = values + dico['hints']
        next_page = dico['_links']['next']['href']
    except KeyError:
        break

listfoodId = []
listlabel = []
listcategory = []
listfoodContentsLabel = []
listimage = []
for i in values:
    keys = i['food'].keys() 
    # Récupérons uniquement les valeurs dont les features demandés sont renseignés :
    if (('foodId' in keys) & ('image' in keys) & ('label' in keys)
            & ('category' in keys) & ('foodContentsLabel' in keys)):
        listimage.append(i['food']['image'])
        listfoodId.append(i['food']['foodId'])
        listlabel.append(i['food']['label'])
        listcategory.append(i['food']['category'])
        listfoodContentsLabel.append(i['food']['foodContentsLabel'])

dic = {}
dic['foodId'] = listfoodId
dic['image'] = listimage
dic['label'] = listlabel
dic['category'] = listcategory
dic['foodContentsLabel'] = listfoodContentsLabel
champagne = pd.DataFrame(dic)

# On ne demande que les 10 premières valeurs : 
champagne.iloc[:10, :].to_csv('champagne.csv', index=False)
champagne.iloc[:10, :]
```
